```python
def drop_useless_columns(df, names):
    return df.drop(names, axis=1)

def convert_xls_to_csv(df, drop_names, rename_names, csv_file_name):
    df = drop_useless_columns(df, drop_names)
    df = df.rename(columns={rename_names[0]: "Year", rename_names[2]: "Destination"})
    for i in range(6, len(rename_names)):
        if ".1" in rename_names[i]:
            col_name = rename_names[i].replace(".1", "_(m)").strip()
            df = df.rename(columns={rename_names[i]:col_name})
        elif ".2" in rename_names[i]:
            col_name = rename_names[i].replace(".2", "_(f)").strip()
            df = df.rename(columns={rename_names[i]:col_name})
        else:
            col_name = (rename_names[i] + "_(mf)").strip()
            df = df.rename(columns={rename_names[i]:col_name})

    df.to_csv(CSV_DIR + '/' + csv_file_name, index=False)
```

```python
def drop_useless_columns(df, names):
    return df.drop(names, axis=1)
```

```python
df_est_refugee_stock = pd.read_excel(XLS_DIR
            + '/total_migrant_stock.xls',
        sheet_name = 'Table 6', header=15)

df_countries_clas = pd.read_excel(XLS_DIR
            + '/total_migrant_stock.xls',
        sheet_name = 'ANNEX', header=15)
```

```python
est_refugee_stock_useless_cols = [
    'Unnamed: 0',
    'Unnamed: 2',
    'Unnamed: 3',
    'Unnamed: 4',
]

est_refugee_stock_cols = list(df_est_refugee_stock.columns)

df_est_refugee_stock = df_est_refugee_stock.rename(columns={
        est_refugee_stock_cols[1]: 'Destination'}
        )

df_est_refugee_stock = drop_useless_columns(
        df_est_refugee_stock,
        est_refugee_stock_useless_cols
        )

for i, col in enumerate(est_refugee_stock_cols):
    if i < 5:
        continue
    
    _col = str(col)

    if '.1' in _col:  # Refugees and asylum seekers as a % of the international migrant stock
        updt_col = _col.replace('.1', '_pct_(mf)').strip()
    elif '-' in _col: # Annual rate of change of the refugee stock (inc. asylum seekers)
        updt_col = _col.strip() + '_annual_rc_(mf)'
    else:            # Refugees and asylum seekers as a % of the international migrant stock
        updt_col = _col.strip() + '_est_(mf)'

    df_est_refugee_stock = df_est_refugee_stock.rename(
            columns={col:updt_col}
            )

df_est_refugee_stock.to_csv(CSV_DIR + '/est_refugee_stock.csv', index=False)
```

```python
countries_clas_useless_cols = [
    'Index',
    'Notes',
    'Code',
    'Type of data',
]

countries_clas_cols = list(df_countries_clas.columns)

df_countries_clas = df_countries_clas.rename(columns={
        countries_clas_cols[1]: 'Destination'}
        )

df_countries_clas = drop_useless_columns(
        df_countries_clas,
        countries_clas_useless_cols
        )

df_countries_clas = df_countries_clas.replace('Y', 1, regex=True)
df_countries_clas.to_csv(CSV_DIR + '/countries_class_by_region.csv', index=False)
```
